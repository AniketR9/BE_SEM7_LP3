```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(x):
  return (x+3)**2
```

```python
def derivative(x):
  return 2*(x+3)
```

```python
start=2
learn_rate=0.01
precision=0.000001
max_iters=10000
step_size=1
x=[]
y=[]
iters=0
x.append(start)
y.append(function(start))
```

```python
while step_size>precision and iters<max_iters:
    prev_x=start
    gradient=derivative(prev_x)
    start=start-learn_rate*gradient
    y.append(function(start))
    x.append(start)
    step_size=abs(start-prev_x)
    print('iteration ',(iters+1),' value = ',start)
    iters+=1

print('The local minima occurs at :',start)
```

```python
plt.scatter(x,y)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('y=(x+3)^2')
plt.show()
```
